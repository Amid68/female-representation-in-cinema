### 3. data types:
- change 'release_date' to 'datetime'

### 4.null values: 
- [] should be considered null value (in the genre, production_companies, and production countries columns). 
- 0.0 should be considered null value in revenue, budget columns.
  

# Cleaning 

```python
# 1.drop columns
movies_df.drop(columns= ['Unnamed: 0', 'year', 'cast_gender', 'crew_gender'], inplace=True)
```

```python
# 2. parse cast and crew
# cast
cast_df = movies_df[['cast']]
cast_df.to_csv('cast_df', index=False)
cast_df.head()
```

```python
cast_df.info()
```

```python
# crew
crew_df = movies_df[['crew']]
crew_df.to_csv('crew_df', index=False)
crew_df.head()
```

```python
crew_df.info()
```

```python
# 3.data tyes
movies_df['bt_score'] = movies_df['bt_score'].astype('category')
movies_df['dubious'] = movies_df['dubious'].astype('category')
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], format='%d/%m/%Y')
```

```python
movies_df.info()
```

```python
# 4.null values
movies_df['revenue'] = movies_df['revenue'].replace({0.0: np.nan})
movies_df['budget'] = movies_df['budget'].replace({0.0: np.nan})
movies_df['genres'] = movies_df['genres'].replace({'[]': np.nan})
movies_df['production_companies'] = movies_df['production_companies'].replace({'[]': np.nan})
movies_df['production_countries'] = movies_df['production_countries'].replace({'[]': np.nan})
movies_df.head()
```

```python
movies_df.head()
```
